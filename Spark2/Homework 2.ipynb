```
filteredRDD=InputRDD.filter(lambda line:len(line.strip())>0)
```

```
filteredRDD.take(2)
```

```
verseid_verse=filteredRDD.map(lambda line: line.split(' ',1))
```

```
Verseid_word_lisRDD=verseid_verse.flatMapValues(lambda x:x.split())
```

```
Verseid_word_lisRDD.collect()
```

```
no=[',', '.', ':', ';', '!', '(', ')', '?']
def word_verseid_kv(value):
  #word_lis=kv_pair[1]
  #verseid=kv_pair[0]
  value=value.lower()
  for i in no:
    if i in value:
      value=value.replace(i,'')
  return value
```

```
word_verseid_kvrdd=Verseid_word_lisRDD.mapValues(word_verseid_kv)
word_verseid_kvrdd.collect()
```

```
reverseRDD=word_verseid_kvrdd.map(lambda kv_pair: (kv_pair[1],kv_pair[0])).distinct()
```

```
reverseRDD.collect()
```

```
def count(kv):
  counts=0
  for i in kv[1]:
    counts+=1
  return (kv[0],kv[1],counts)
  
  
```

```
count_verseidRDD=word_verseidlisRDD2.map(count)
count_verseidRDD.collect()
```
